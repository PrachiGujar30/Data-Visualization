```python
import pandas as pd
import seaborn as sns
d=pd.read_excel("/content/super2.xlsx")
d.head()
```

```python
import plotly.express as px
fig = px.treemap(d, path=["Category","Sub-Category"],
                 values='Sales')
fig.show()
```

```python
sns.barplot(x='Category',y='Sales',data=d,palette='plasma',orient='v')
```

```python
sns.barplot(x='Sales',y='Segment',data=d,palette='plasma',
            orient='h',hue='Category',estimator=sum)
```

```python
sns.catplot(y='Sales',x='Segment',data=d,palette='summer',
            orient='v',hue='Category',estimator=sum,
            order=['Corporate','Consumer','Home Office'],kind='bar')
```

```python
sns.lmplot(data=d,x='Sales',y='Profit',hue='Segment')
```

```python
sns.distplot(d.Profit)
```

```python
sns.distplot(d.Profit, bins=5, kde=False, color='r')
```

```python
sns.kdeplot(d['Sales'], fill=True)
```
