```python
#DEALING WITH MISSING VALUES
import pandas as pd
import numpy as np

df=pd.read_excel("/content/Emp.xlsx")

CountofEmp = 0
avg = 0

for CGPA in df['CGPA']:
  if str(CGPA).replace('.','',1).isdigit():
    CountofEmp += 1
    avg += CGPA
avg /= CountofEmp

df = df.replace(to_replace=np.nan, value=avg)

df
```
